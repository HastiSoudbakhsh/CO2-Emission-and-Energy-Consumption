```python
# the 'Renewable energy share in the total final energy consumption (%)' in all countries for the year 2020 is empty.
# also for the country 'Albania' all the column is empty.

# for solving first solution I define a function that for all countries use the mean of last 3 years for filling out the missing values
# for second solution there is 2 approach: 1-linear regression 2-because this country is mostly look like 'georgia', I can use Georgia's values


b = data

def fill_missing_values(b):
    for country in b['Entity'].unique():
        country_data = b[b['Entity'] == country]

        missing_values = country_data['Renewable energy share in the total final energy consumption (%)'].isnull()

        # If there are missing values, fill only those with the mean of Energy Share for 2017, 2018, and 2019
        if missing_values.any():
            missing_indices = missing_values[missing_values].index
            mean_energy_share = country_data.loc[country_data['Year'].isin([2017, 2018, 2019]),
                                                 'Renewable energy share in the total final energy consumption (%)'].mean()
            b.loc[missing_indices, 'Renewable energy share in the total final energy consumption (%)'] = mean_energy_share

    return b


df_filled = fill_missing_values(b)
```

```python
b = data.copy()

def fill_missing_values(b):
    for country in b['Entity'].unique():
        country_data = b[b['Entity'] == country]

        missing_values = country_data['Value_co2_emissions_kt_by_country'].isnull()

        if missing_values.any():
            missing_indices = missing_values[missing_values].index
            last_year_value = country_data.loc[country_data['Year'] == 2019, 'Value_co2_emissions_kt_by_country'].values

            # Check the condition: if year[2019] > year[2018], use forward fill; else, use the mean of the last year
            if last_year_value > country_data.loc[country_data['Year'] == 2018, 'Value_co2_emissions_kt_by_country'].values:
                # Use forward fill for missing values
                b.loc[missing_indices, 'Value_co2_emissions_kt_by_country'] = country_data['Value_co2_emissions_kt_by_country'].ffill()
            else:
                # Use the mean of the last year for filling missing values
                mean_energy_share = country_data.loc[country_data['Year'].isin([2017, 2018, 2019]), 'Value_co2_emissions_kt_by_country'].mean()
                b.loc[missing_indices, 'Value_co2_emissions_kt_by_country'] = mean_energy_share

    return b

filled_data = fill_missing_values(b)
```
